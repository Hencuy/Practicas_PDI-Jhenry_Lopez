```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('Pelon.jpeg')
gris = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

_,mascara = cv2.threshold(gris,127,255,cv2.THRESH_BINARY)

#Matriz para poder erosinar
kernel = np.array([[0,1,0],[1,1,1],[1,1,1]],dtype=np.uint8) #personalizar

#Realiza la erosion
erosion = cv2.erode(img,kernel,iterations=30) 

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagen Erosinada')
plt.imshow(erosion, cmap='gray')
plt.axis('off')

plt.show()




```

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('Pelon.jpeg')
img_2 = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

arra1 = np.array([145,230,0])
arra2 = np.array([180,255,255])

copia = img_2.copy()

mascara = cv2.inRange(copia,arra1,arra2)

kernel = np.ones((5,5),np.uint8)

erosion = cv2.erode(mascara,kernel, iterations=3)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagen Erosinada')
plt.imshow(erosion, cmap='gray')
plt.axis('off')

plt.show()
```

Circulos Aleatorios

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

canvas = np.zeros((350, 500, 3), dtype='uint8')

for i in range(0, 5):
    radius = np.random.randint(5, high=100)
    color = np.random.randint(0, high=256, size=(3,)).tolist()

    pt = np.random.randint(0, high=300, size=(2,))

    cv2.circle(canvas, tuple(pt), radius, color, -1)


# Display original image
plt.subplot(1, 2, 1)
plt.imshow(canvas, cmap='gray')
plt.title('Original Image')




plt.show()


```

Triangulo 

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

img = cv2.imread('Monin.jpeg', cv2.IMREAD_GRAYSCALE)

alto, ancho = img.shape[:2]

mascara = np.zeros_like(img)
circulo = (ancho // 2, alto // 3)

radio = 300
cv2.circle(mascara, circulo, radio, (255), -1)

img_fin = cv2.bitwise_and(img, img, mask=mascara)

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img_fin, cmap='gray')
plt.title('img_tras')

plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np


img = cv2.imread('Monin.jpeg', cv2.IMREAD_GRAYSCALE)

alto, ancho = img.shape[:2]

mascara = np.zeros_like(img)

#Puntos del triangulo 
a = ancho // 2, alto // 3
b = ancho // 4, alto * 2 // 3
c = ancho * 2 // 3, alto * 2 // 3


puntos_triangulo = np.array([[a], [b], [c]])


cv2.fillPoly(mascara, [puntos_triangulo], 255)

img_fin = cv2.bitwise_and(img, img, mask=mascara)



# Display original image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(img_fin, cmap='gray')
plt.title('Imagen con Triangulo')

plt.show()
```

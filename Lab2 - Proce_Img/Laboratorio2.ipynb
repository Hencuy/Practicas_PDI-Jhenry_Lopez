```python
import matplotlib.pyplot as plt 
import cv2 
import numpy as np 

imagen = 'img1.jpg'
#Primer paso
imagenBGR = cv2.imread( imagen ,cv2.IMREAD_COLOR)

#Segundo paso
imagenRGB = cv2.cvtColor(imagenBGR, cv2.COLOR_BGR2RGB)

#Tercer paso
imagengris = cv2.cvtColor(imagenBGR, cv2.COLOR_BGR2GRAY)


#Cuarto paso
cv2.imwrite('img1Guardado.jpg',imagengris)

#Quinto paso
imgguardada = cv2.imread('img1Guardado.jpg', cv2.IMREAD_UNCHANGED)


#Sexto paso
def mostrar_imagenes(imagenRGB, imgguardada):
    plt.figure()
    plt.subplot(121)
    plt.title('Original')
    plt.imshow(imagenRGB)

    plt.subplot(122)
    plt.title('En escala de grises')
    plt.imshow(imgguardada, cmap='gray')
    
    plt.show()


mostrar_imagenes(imagenRGB, imgguardada)



```

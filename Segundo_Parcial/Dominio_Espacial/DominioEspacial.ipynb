Convolucion 
Es el tratamiento de una matriz por otra que se llama Kernel
La convolucion es simplemente una suma pesada de pixeles vecinos(3x3) del pixel fuente


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

img = cv2.imread('Pelon.jpeg',0)
mascara = np.array([[1,-1,1],[-1,2,-1],[1,-1,1]])

imgfin = cv2.filter2D(img,-1,mascara)

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(122)
plt.imshow(imgfin, cmap='gray')
plt.title('Convolucion')
plt.axis('off')
```

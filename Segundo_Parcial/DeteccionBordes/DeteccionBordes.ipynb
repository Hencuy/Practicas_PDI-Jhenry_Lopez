```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('fullgays.jpeg', 0)

mascaraRobertsX = np.array([[1, 0], [0, -1]])
mascaraRobertsY = np.array([[0, 1], [-1, 0]])

mascaraPrewittX = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
mascaraPrewittY = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

mascaraSobelX = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
mascaraSobelY = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

RobertsX = cv2.filter2D(img, -1, mascaraRobertsX)
RobertsY = cv2.filter2D(img, -1, mascaraRobertsY)
Roberts = np.sqrt(np.square(RobertsX) + np.square(RobertsY))

PrewittX = cv2.filter2D(img, -1, mascaraPrewittX)
PrewittY = cv2.filter2D(img, -1, mascaraPrewittY)
Prewitt = np.sqrt(np.square(PrewittX) + np.square(PrewittY))

SobelX = cv2.filter2D(img, -1, mascaraSobelX)
SobelY = cv2.filter2D(img, -1, mascaraSobelY)
Sobel = np.sqrt(np.square(SobelX) + np.square(SobelY))

bordes_canny = cv2.Canny(img, 100, 200)

plt.figure(figsize=(15, 8))
plt.subplot(321)
plt.title('Original')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(322)
plt.title('Roberts')
plt.imshow(Roberts, cmap='gray')
plt.axis('off')

plt.subplot(323)
plt.title('Prewitt')
plt.imshow(Prewitt, cmap='gray')
plt.axis('off')

plt.subplot(324)
plt.title('Sobel')
plt.imshow(Sobel, cmap='gray')
plt.axis('off')

plt.subplot(325)
plt.title('Bordes canny')
plt.imshow(bordes_canny, cmap='gray')
plt.axis('off')

plt.show()
```

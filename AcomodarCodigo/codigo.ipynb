Codigo Ordenado que comprende binarios, inversa binaria, umbralizacion adaptativa y umbralizacion gaussiana

```python
import cv2 
import matplotlib.pyplot as plt

imagen = cv2.imread('llama.jpg',0)

def histograma(imagen):
    histo = cv2.calcHist([imagen],[0],None,[256],[0,256])
    plt.figure(figsize=(10,5))
    plt.plot(histo,color="black")
    plt.title("histograma")
    plt.xlabel('intencidad_pixel')
    plt.ylabel('frecuencia_pixel')
    plt.show()



#Esta es la umbralizacion binaria en binario solo se trabajo entre 0  y 255
_,umbralbi = cv2.threshold(imagen,127,255,cv2.THRESH_BINARY)
#Invertimos el binario
_,Umbral_inve_bi = cv2.threshold(imagen,127,255,cv2.THRESH_BINARY_INV)

#Umbralizacion Adaptativa
umb_adap = cv2.adaptiveThreshold(imagen,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)

#Umbralizacion Gausiana
umb_adap_gauss = cv2.adaptiveThreshold(imagen,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
#histograma
histo = cv2.calcHist([imagen],[0],None,[256],[0,256])


plt.figure(figsize=(10,5))

#Referencia al array de las imagenes
plt.subplot(2,3,1)
plt.imshow(imagen,cmap='gray')
plt.title('Imagen incial')


plt.subplot(2,3,2)
plt.imshow(umbralbi,cmap='gray')
plt.title('Imagen binaria')

plt.subplot(2,3,3)
plt.imshow(Umbral_inve_bi,cmap='gray')
plt.title('Imagen Inversa binaria')



plt.subplot(2,3,4)
plt.imshow(umb_adap,cmap='gray')
plt.title('Umbralizacion adaptativa')

plt.subplot(2,3,5)
plt.imshow(umb_adap_gauss,cmap='gray')
plt.title('Umbralizacion Gauss')

histograma(imagen)

```
